```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\mohan\Downloads\gender_submission.csv"
data = pd.read_csv(file_path)
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

```python
print(data.isnull().sum())
```

```python
print("Column Names:", data.columns)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='Survived')
plt.title('Distribution of Survival')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='PassengerId', hue='Survived')
plt.title('Survival Count by PassengerId')
plt.xlabel('PassengerId')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
survival_rates = data.groupby('PassengerId')['Survived'].mean().reset_index()
print(survival_rates)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(data=survival_rates, x='PassengerId', y='Survived')
plt.title('Survival Rate by PassengerId')
plt.xlabel('PassengerId')
plt.ylabel('Survival Rate')
plt.xticks(rotation=90)
plt.show()
```
